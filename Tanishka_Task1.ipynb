##classes & OOP

```python
# Write a class in Python for a point (1D)
# Use functions like __str__ for it. Find out what they do and write down some other dunder functions and what they do.

class Point1D:
  def __init__(self,x=0):
    self.x=x
  def __str__(self):
    return f"(point : {self.x})"
  def __eq__(self,other):
    if isinstance(other,Point1D):
      return self.x == other.x
    return False
  def __add__(self,other):
    if isinstance(other,Point1D):
      return Point1D(self.x + other.x)
    return NotImplemented
  def __lt__(self,other):
    return self.x < other.x
  def __gt__(self,other):
    return self.x > other.x
  def dist_from_origin(self):
    return abs(self.x)
p1 = Point1D(2)
p2 = Point1D(6)

print(p1)
print(p1 == p2)
print(p1 + p2)
print(p2 < p1)
print(p2 > p1)

# Extend it for a Line. (2D)
class Point2D:
  def __init__(self,x,y):
    self.x =x
    self.y =y
  def __str__(self):
    return f"point ({self.x},{self.y})"
  def __add__(self,other):
    if isinstance(other,Point2D):
      return Point2D(self.x + other.x, self.y + other.y)
    return NotImplemented
  def __sub__(self, other):
    if isinstance(other, Point2D):
        return Point2D(self.x - other.x, self.y - other.y)
    return NotImplemented

class Line2D:
  def __init__(self,p1,p2):
    if not isinstance(p1, Point2D) or not isinstance(p2, Point2D):
          raise TypeError("Line endpoints must be Point2D objects")
    self.p1 = p1
    self.p2 = p2
  def __str__(self):
    return f"Line({self.p1}, {self.p2})"
  def length(self):
    return ((self.p2.x - self.p1.x)**2 + (self.p2.y -self.p1.y)**2)**(0.5)
  def slope(self):
    if self.p2.x == self.p1.x:
      return float('inf')
    return (self.p2.y - self.p1.y) / (self.p2.x - self.p1.x)
p1 = Point2D(0, 0)
p2 = Point2D(3, 4)
p3 = Point2D(3, 4)

line1 = Line2D(p1, p2)
line2 = Line2D(p2, p1)
line3 = Line2D(p1, p3)

print(p1)
print(p1+p2)
print(p2-p1)

print(line1)
print(line2)
print(line3)

print(line1.length())
print(line1.slope())
print(line2.length())
print(line2.slope())
print(line3.length())
print(line3.slope())
print(line1 == line2)
```

### Univariate Analysis

Univariate analysis focuses on analyzing each feature in the dataset independently.

- **Distribution analysis**: The distribution of each feature is examined to identify its shape, central tendency, and dispersion.

- **Identifying potential issues**: Univariate analysis helps in identifying potential problems with the data such as outliers, skewness, and missing values

#### The shape of a data distribution refers to its overall pattern or form as it is represented on a graph. Some common shapes of data distributions include:

- **Normal Distribution**: A symmetrical and bell-shaped distribution where the mean, median, and mode are equal and the majority of the data falls in the middle of the distribution with gradually decreasing frequencies towards the tails.

- **Skewed Distribution**: A distribution that is not symmetrical, with one tail being longer than the other. It can be either positively skewed (right-skewed) or negatively skewed (left-skewed).

- **Bimodal Distribution**: A distribution with two peaks or modes.

- **Uniform Distribution**: A distribution where all values have an equal chance of occurring.

The shape of the data distribution is important in identifying the presence of outliers, skewness, and the type of statistical tests and models that can be used for further analysis.

#### **Dispersion** is a statistical term used to describe the spread or variability of a set of data. It measures how far the values in a data set are spread out from the central tendency (mean, median, or mode) of the data.

There are several measures of dispersion, including:

- **Range**: The difference between the largest and smallest values in a data set.

- **Variance**: The average of the squared deviations of each value from the mean of the data set.

- **Standard Deviation**: The square root of the variance. It provides a measure of the spread of the data that is in the same units as the original data.

- **Interquartile range (IQR)**: The range between the first quartile (25th percentile) and the third quartile (75th percentile) of the data.

Dispersion helps to describe the spread of the data, which can help to identify the presence of outliers and skewness in the data.

## Why KDE and Box Plot though??

We draw a **boxplot** to quickly see the spread of data, its median, and especially to detect **outliers**, since it neatly summarizes the distribution in a compact form, while we draw a **KDE plot** (Kernel Density Estimate) to understand the **shape of the distribution** smoothly, which helps us see whether the data is normal, skewed, or has multiple peaks.

**In short:**
- **Boxplots** are best for spotting outliers and comparing groups
- **KDE plots** are best for visualizing the overall distribution pattern

# Feature Scaling :

Essential data pre-processing technique used to standardize the range of independent variables or features. Since ML algorithms often rely on the relative magnitude of values, scaling ensures that no single feature dominates others due to its range.

two primary methods of feature scaling:
1. Normalisation
2. Standardisation

---

# Standardization
  - What is standardisation ?
    * It is used to bring uniformity to the datasets , resulting in independent variables and features of the same scale and range.

    * Transforms the standard deviation to 1 and the mean to 0

  - In standardization, the mean is subtracted from each data point and the result obtained is divided by the standard deviation , resulting in standardized and rescaled data.

> Formula :
        
  ## Z = (X - m ) / s
        X = data point
        m = mean
        s = std deviation






> # Code Implementation of Z Score Normalisation











```python
print(df['Age'])
```

```python
from sklearn.preprocessing import StandardScaler #applies the formula directly
```

```python
# object will learn the mean and standard deviation of the Age column.
scaler = StandardScaler()
```

```python
# Handle missing values in 'Age' by filling with mean
df['Age'].fillna(df['Age'].mean(), inplace=False)  # Avoid inplace to prevent warnings; assign instead
df['Age_Filled'] = df['Age'].fillna(df['Age'].mean())
```

```python
#fit : calc mean and std dev
#transform : applies z score transfrmation
#[['Age']] : passed as a 2D array
df['Age_standardization']=scaler.fit_transform(df[['Age_Filled']])
```

```python
#comparitive study before and after scaling
print(df[['Age', 'Age_standardization']].head(100))
```



> Manual Implementation





```python
# Step 1: Calculate mean of Age
mean_age = df['Age'].mean()
# Step 2: Calculate standard deviation of Age
std_age = df['Age'].std()
# Step 3: Apply Z-score formula manually
df['Age_Standardized'] = (df['Age'] - mean_age) / std_age
print(df['Age_Standardized'])
```



> Verification


```python
print("Mean before scaling:", round(df['Age'].mean()))
print("Std before scaling:", round(df['Age'].std()))
print("Mean after scaling:",round( df['Age_standardization'].mean()))
print("Std after scaling:", round(df['Age_standardization'].std()))
```

# Normalization

 required only when your dataset has features of varying ranges

  - What is normalisation?
    * Resizes the range of feature values to a specific scale, usually between 0 and 1
    * Feature scaling technique used to transform data into a standard range.
    * Normalization ensures that features with different scales or units contribute equally to the model and improves the performance of many machine learning algorithms.

  - Key Features:
    * Maps the minimum and maximum of a feature to a defined range
    * Preserves the relative relationships of the original data
    * Useful for algorithms that rely on distance metrics such as k-Nearest Neighbours and clustering



> Min-Max Scaling Formula :
# X_normalized = X – X_min / X_max – X_min
        X = random feature value
        X_min = = min feature val in data
        X_max = max feature val in data

#Difference between Standardisation and Normalization

While both are feature scaling techniques, they differ in their approaches and applications. Understanding these differences is key to choosing the right technique for your machine learning model.

* Rescaling Methods
  - Normalization rescales feature values within a predefined range, often between 0 and 1
  - Particularly useful for models where the scale of features varies greatly.
  - while , Standardization centers data around the mean (0) and scales it according to the standard deviation (1).

* Sensitivity to Outliers
  - Normalization techniques do not handle the outlier problem as effectively as standardization
  - As standardization explicitly relies on both the mean and the standard deviation.

* Use cases
  - Normalization
    + distance-based algorithms
      1.  k-Nearest Neighbors (k-NN)
    
    to prevent features with larger scales from dominating the distance calculations.

    + Unknown or Non-Gaussian Distribution
      
    When the distribution of data is not known or does not follow a normal (Gaussian) pattern.

  - Standardization
  
  prevents feature dominance in optimization and variance-based techniques.
    + Gradient-based Algorithms
      1. SVM
      2. Logistic Regression
      3. Linear Regression
    + Dimentionality Reduction
      1. PCA

---

#Pros and Cons

Pro:

  - The advantages include improved model performance and balanced feature contributions.

Con:

  - Normalization can limit interpretability due to the fixed scale
  - Standardization can also make interpretation harder as the values no longer reflect their original units.


There’s always a trade-off between model complexity and model accuracy.
